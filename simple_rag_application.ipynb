```python
import numpy as np
```

```python
def embedding_model(query, openai_client, model="text-embedding-3-small"):
    # This function calculates and returns an embedding
    response = openai_client.embeddings.create(
        model = model,
        input = query
    )
    return response.data[0].embedding
```

```python
def cosine_similarity(vec1, vec2):
    # This function creates cosine similarity using only numpy
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
```

```python
vector_database = {}

for q in faq_database:
    vector_database[q] = embedding_model(q, client)
```

```python
def find_most_similar_faq(query, faq_vector_db, openai_client):
    query_embedding = embedding_model(query, openai_client)
    best_match = None
    highest_similarity = -1

    for faq_question, faq_embedding in faq_vector_db.items():
        similarity = cosine_similarity(query_embedding, faq_embedding)
        if similarity > highest_similarity:
            best_match = faq_question
            highest_similarity = similarity
    
    return best_match
```

```python
user_query_1 = "Can I return a product after 30 days?"
user_query_2 = "How much time do I have to return my product?"
user_query_3 = "Where can I follow my order?"

best_match = find_most_similar_faq(user_query_3, vector_database, client)

print(f"The most similar FAQ to your query is: '{best_match}'")
print(f"Answer: {faq_database[best_match]}")
```
